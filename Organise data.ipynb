# Preprocess IXData

```python
IXdatapath = './data/IXData'
with open('textfiles/filepaths.txt') as f:
    filepaths_raw = np.array(f.read().splitlines())
    filepaths = np.array([x.split('/') for x in filepaths_raw])

lengths = np.array([len(x) for x in filepaths])
imageIDs = np.array([x[3] + re.findall('_\D\d+_s\d_w\d',x[-1])[0] for x in filepaths])
IDcounts_dict = dict(zip(pd.Series(['_'.join(x.split('_')[0:2]) for x in imageIDs]).value_counts().index, list(pd.Series(['_'.join(x.split('_')[0:2]) for x in imageIDs]).value_counts())))
IDcounts = np.array([IDcounts_dict['_'.join(x.split('_')[0:2])] for x in imageIDs])

cleanIDs_index = np.bitwise_and((IDcounts == 8), (lengths == 7))
print sum(IDcounts != 8), "ID with more than 8 replicates"
print sum(lengths != 7), "ID without a plateID"
dirtyIDs_index = [not x for x in np.bitwise_and((IDcounts == 8), (lengths == 7))]
print sum(dirtyIDs_index), "total dirty IDs"

# 8 copies of each image ID, 4 replicates, 2 channels.
clean_IDs = imageIDs[cleanIDs_index]
all(np.array(list(pd.Series(['_'.join(x.split('_')[0:2]) for x in clean_IDs]).value_counts())) == 8) ## True

clean_filepaths_raw = filepaths_raw[cleanIDs_index]

## Build mapping from Run to plate ID
RunIds_dict = {}
for p in [(x[1], x[3]) for x in filepaths]:
    if p[0] in RunIds_dict:
        RunIds_dict[p[0]].add(p[1])
    else:
        RunIds_dict[p[0]] = set([p[1]])
        
cleanimageIDs = imageIDs[cleanIDs_index]

for (i,p) in enumerate(clean_filepaths_raw):
    if i % 100 == 0:
        print i
    shutil.copy('./data/IXData/' + p, './processed/IXClean' +  cleanimageIDs[i] + '.tif')
```

# Preprocess Positive IDs

```python
with open('textfiles/PositiveControllist.txt') as f:
    rawpositiveIDs = f.read().splitlines()

cleanpositiveIDs = ['_'.join([x.split('_')[0], x.split('_')[2], x.split('_')[3], x.split('_')[4][0:2]]) for x in rawpositiveIDs]

for (i,p) in enumerate(cleanpositiveIDs):
    shutil.copy('./data/Positive Controls/' + rawpositiveIDs[i], './processed/CleanPositiveControls/' + p + '.tif')
```

# Preprocess Negative IDs

```python
with open('textfiles/NegativeControllist.txt') as f:
    rawnegativeIDs = f.read().splitlines()

cleannegativeIDs = ['_'.join([x.split('_')[0], x.split('_')[2], x.split('_')[3], x.split('_')[4][0:2]]) for x in rawnegativeIDs]

for (i,p) in enumerate(cleannegativeIDs):
    shutil.copy('./data/Negative Controls/' + rawnegativeIDs[i], './processed/CleanNegativeControls/' + p + '.tif')
```
